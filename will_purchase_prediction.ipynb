```python
import pickle
import pandas as pd
```

```python
with open('knn_model.pickle','rb') as f:
  knn_new=pickle.load(f)
```

```python
with open('scaler.pickle','rb') as f:
  scaler_new=pickle.load(f)
```

```python
new_df=pd.read_csv('/content/new_customers.csv')
```

```python
new_df
```

```python
gender_encoded_new=pd.get_dummies(new_df['Gender'], drop_first=True)
```

```python
gender_encoded_new
```

```python
df_new2= pd.concat([new_df,gender_encoded_new], axis=1)
```

```python
df_new2
```

```python
X_new=df_new2[['Male','Age','Salary','Price']].to_numpy()
```

```python
X_new
```

```python
X_new_scaler=scaler_new.fit_transform(X_new)
```

```python
X_new_scaler
```

```python
y_new_pred= knn_new.predict(X_new_scaler)
```

```python
y_new_pred
```

```python
df_new2['will_purchase']=y_new_pred
```
